### Calculating Outliers,Skewness & Kurtosis 

#### Importing Necessary Libraries:

```python
# Importing scipy 
import scipy
from scipy.stats import skew
from scipy.stats import kurtosis
import seaborn as sns
```

```python
# Outliers Detection using IqR (Only for numeric columns) 
numeric_cols=final_data.select_dtypes(include=['number']).columns
Q1=final_data[numeric_cols].quantile(0.25)
Q3=final_data[numeric_cols].quantile(0.75)
IQR=Q3-Q1

outliers = ((final_data[numeric_cols] < (Q1 - 1.5 * IQR)) | (final_data[numeric_cols] > (Q3 + 1.5 * IQR))).sum()
print("Outliers Count Per Column:\n", outliers)
```

```python
# Calculate the skewness 
numeric_cols_excluding_datetime = final_data.select_dtypes(include=['number']).columns
df_skew = final_data[numeric_cols_excluding_datetime].skew()
df_skew.drop(labels=['season','yr'], errors='ignore', inplace=True)
print("Skewness: \n", df_skew)
```

```python
# Correlation matrix (excluding categorical columns)
numeric_features=final_data.select_dtypes(include=['number']).columns
df_corr = final_data[numeric_features].corr()
print("Correlation Matrix:\n", df_corr)
```

```python
#Compute correlation matrix
numeric_features = final_data.select_dtypes(include=['number']).columns
df_corr = final_data[numeric_features].corr()

#Drop columns that contain only NaN volues in the correlation matrix
df_corr_cleaned = df_corr.dropna(how='all', axis=1).dropna(how ='all', axis=0)
print("Cleaned Correlation Matrix:\n", df_corr_cleaned)
```
